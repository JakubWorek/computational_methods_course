# Zadanie 1 - Metoda potęgowa
Napisz funkcję obliczającą metodą potęgową dominującą wartość własną (największą co  
do modułu) i odpowiadający jej wektor własny dla danej macierzy rzeczywistej syme-  
trycznej. Sprawdź poprawność działania programu porównując własną implementację  
z wynikami funkcji bibliotecznej. Przedstaw na wykresie zależność czasu obliczeń od  
rozmiaru macierzy (rozmiary macierzy 100x100, 500x500, ...).  
•Powtarzaj mnożenie wektora xi przez macierz A:  
xi+1 = Axi,  
dzieląc za każdym razem wektor wynikowy przez ||xi+1||∞  
•Element wektora xi o największej wartości bezwzględnej zbiega do dominującej  
wartości własnej  
•Przeskalowany wektor xi zbiega do dominującego wektora własnego  
•Obliczenia powinny się zatrzymać po przekroczeniu maksymalnej liczby iteracji,  
albo w przypadku gdy ||xi −xi+1||< ε (kryterium małej poprawki)  
•Pod koniec obliczeń znormalizuj otrzymany wektor własny.  

```python
def metoda_potegowa(A, eps=1e-6, maxIt=10000):
    n = len(A)
    v_k = np.array([1/n] * n)
    delta = 1

    i = 0
    while delta > eps and i < maxIt:
        v_kk = A @ v_k
        lambda_k = np.linalg.norm(v_k, np.inf)
        v_kk = v_kk / lambda_k
        delta = np.linalg.norm(v_k - v_kk)
        v_k = v_kk
        i += 1
    return v_k / np.linalg.norm(v_k), i
```

```python
def metoda_potegowa_lib(A):
    return np.linalg.eig(A)[1].transpose()[0]
```

```python
A_copy = deepcopy(A)
eps = 1e-6
max_iter = 10000

start = time.time()
print(metoda_potegowa(A, eps, max_iter))
end = time.time()
print(end - start)
```

# Zadanie 2 - Odwrotna metoda potęgowa
Opierając się na twierdzeniu o transformacji widma macierzy:  
Twierdzenie 1 Macierz (A−σI)−1 (jeśli istnieje), to ma wartości własne równe 1 /   
λk −σ  
(λk jest k-tą wartością macierzy A) i wektory własne identyczne z macierzą A.  
oraz wykorzystując metodę potęgową i faktoryzację LU zaimplementuj odwrotną metodę  
potęgową pozwalającą na szybkie znalezienie wektorów własnych macierzy A, dla warto-  
ści σ bliskich odpowiedniej wartości własnej. Wykorzystaj fakt, że mnożenie wektora xi  
przez macierz A−1 (xi+1 = A−1xi) odpowiada rozwiązaniu układu równań Axi+1 = xi.  

```python
def odwrotna_metoda_potegowa(A, eps=1e-5, maxIt=10000, sigma=1):
    n = len(A)
    v_k = np.array([1/n] * n)
    B = linalg.inv(A - sigma * np.eye(n))
    delta = 1
    
    lu = linalg.lu_factor(B)
    
    i = 0
    while delta > eps and i < maxIt:
        v_kk = linalg.lu_solve(lu, v_k)
        v_kk = v_kk / np.linalg.norm(v_kk)
        delta = np.linalg.norm(v_k - v_kk)
        v_k = v_kk
        i += 1

    return v_k / np.linalg.norm(v_k), i
```

```python
A_copy = deepcopy(A)
eps = 1e-6
max_iter = 10000
sigma = 0

start = time.time()
print(odwrotna_metoda_potegowa(A, eps, max_iter, sigma))
end = time.time()
print(end - start)
```

# Zadanie 3 - Iteracje z ilorazer Rayleigha
Zaimplementuj iteracyjną metodę wyznaczania wartości własnej i skojarzonego z nią  
wektora własnego wykorzystując odwróconą metodę potęgową oraz iloraz Rayleigha.  
Porównaj zbieżność metody ze zbieżnością algorytmu potęgowego (macierz symetryczna  
rzeczywista).  
r(x) = xAxT / xxT  
r(qi) = λi

```python
def raileigh(x, A):
    return (x.transpose() @ A @ x) / (x.transpose() @ x)

def iloraz_rayleigha(A, eps, maxIt, sigma):
    lambda_k = sigma
    n = len(A)
    v_k = np.array([1/n] * n)
    delta = 1

    i = 0
    while delta > eps and i < maxIt:
        B = A - lambda_k * np.eye(n)
        v_kk = np.linalg.solve(B, v_k)
        v_kk = v_kk / np.linalg.norm(v_kk)
        lambda_k = raileigh(v_k, A)
        delta = np.linalg.norm(v_k - v_kk)
        v_k = v_kk

    return v_k / np.linalg.norm(v_k), i
```

```python
A_copy = deepcopy(A)
eps = 1e-6
max_iter = 10000
sigma = 0

start = time.time()
print(iloraz_rayleigha(A, eps, max_iter, sigma))
end = time.time()
print(end - start)
```
